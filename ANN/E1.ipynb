```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data = pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'],axis =1)
```

```python
data.head()
```

```python
leg = LabelEncoder()
data['Gender'] = leg.fit_transform(data['Gender'])
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Create and fit the encoder
ohe = OneHotEncoder()
values = ohe.fit_transform(data[['Geography']]).toarray()

# Convert to DataFrame using the fitted encoder
ohe_df = pd.DataFrame(values, columns=ohe.get_feature_names_out(['Geography']))
```

```python
ohe_df
```

```python
data = data.drop(['Geography'],axis =1)
data = pd.concat([data,ohe_df],axis = 1)
data
```

```python
with open('leg.pkl','wb') as file:
    pickle.dump(leg,file)

with open('ohe.pkl','wb') as file:
    pickle.dump(ohe,file)
```

```python
x = data.drop(['Exited'],axis = 1) ##split the data into dependant and independant variables (x and y)
y = data['Exited']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42) ## basically again splitting it 80 percent for training and 20 percent for testing.

scaler = StandardScaler() ## the scaler fucntion basically finds the mean and standard devaition.
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
##using this pickle code we can save our trained objs(for example here we are storing scaler obj in this file which can be used again if needed)
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
```

```python
model = Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),  ##here x-train.shape(1) basically return the number of parametrs in x_train.
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid') ##output layer 
])
```

```python
model.summary() ## basically gives info about the model we created .
```

```python
import tensorflow
opt =  tensorflow.keras.optimizers.Adam(learning_rate = 0.01) ## we are defining an optimiser function namely Adam that basically is used for updating wieghts and ultimately reducing the loss and learning rate controls how fast the optimiser learns. 
loss = tensorflow.keras.losses.BinaryCrossentropy() ## calculates the loss.
model.compile(optimizer=opt,loss=loss,metrics=['accuracy']) ## now we are compiling the model and metrics basically gives back some parameters of the model after each step(here it returns accuracy.).
```

```python
import datetime
log_dir = 'logs/fit/' + datetime.datetime.now().strftime("%Y%m%d-%H%M%S") 
## if log directories are created without date and time and new log will overwrite the old log.
## datetime.datetime.now().strftime("%Y%m%d-%H%M%S") this basically return the current date and time but in the form of a sting.
tc = TensorBoard(log_dir=log_dir,histogram_freq=1)

## histogram_freq=1 ==> Log histograms of weights and activations once every epoch (Histograms show the distribution of weights, biases, and activations inside each layer over time.)
```

```python
esc = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
## This basically monitors the loss and if the loss doesnt decrease after 5 straight epochs then it stops even though more epochs were alloted.
## After stopping it returns the model with the least loss value. 
## the monitor can be used to study the validation loss or training loss or validation accuracy or training accuracy.
```

```python
## Training the model

history = model.fit(
    x_train,y_train, ## This data will be used for training.
    validation_data=(x_test,y_test), ## This data will be used for comparing and returning loss and other metrics 
    epochs = 100, ## run for 100 epochs
    callbacks = [esc , tc] ## use these callbacks as well.
)
```

```python
model.save('model.h5') ##saves the entire model in HDF5 format.

## tensorboard --logdir=logs/fit ==> use this command in the terminal to launch the tensorboard in the webbrowser.
```
